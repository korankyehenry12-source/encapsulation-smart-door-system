## Smart Door Access Control System (UMaT)



```python
class Staff:
    """
    UMaT Staff Access Card — Encapsulated Design
    """

    def __init__(self, s_name, access_code):
        self.s_name = s_name
        self.__access_code = access_code
        self.__pin_entered = False

    def enter_pin(self, pin):
        """
        Enter the admin PIN to unlock access code viewing.
        """
        if pin == "1234":
            self.__pin_entered = True
            print(f"🔓 {self.s_name}'s access code is now visible.")
            return True
        print(f"🔒 Incorrect PIN for {self.s_name}.")
        return False

    @property
    def access_code(self):
        """
        Get the access code. Requires PIN entry first.
        """
        if not self.__pin_entered:
            return "[ACCESS BLOCKED — Enter admin PIN first]"
        return self.__access_code

    @access_code.setter
    def access_code(self, value):
        """
        Set a new access code. Must be 4 or more characters.
        """
        if len(value) < 4:
            raise ValueError(f"Code '{value}' rejected: minimum length is 4.")
        self.__access_code = value
        print(f"✅ {self.s_name}: New access code saved.")

    def display_staff_info(self):
        """
        Print staff details with hidden access code.
        """
        mask = "•" * len(self.__access_code)
        print("─" * 38)
        print(f"  UMaT STAFF ACCESS RECORD")
        print("─" * 38)
        print(f"  Name: {self.s_name}")
        print(f"  Code: {mask}")
        print(f"  PIN Status: {'Unlocked' if self.__pin_entered else 'Locked'}")
        print("─" * 38)
```

## Running the System

```python
staff = Staff("Dr. Matthew Cobbinah", "UMAT54")
staff.display_staff_info()

print("\n--- Without PIN ---")
print(staff.access_code)

print("\n--- Wrong PIN ---")
staff.enter_pin("0000")

print("\n--- Correct PIN ---")
staff.enter_pin("1234")
print(f"Code: {staff.access_code}")

print("\n--- Invalid update ---")
try:
    staff.access_code = "XY"
except ValueError as e:
    print(f"Error: {e}")

print("\n--- Valid update ---")
staff.access_code = "GATE2024"
staff.display_staff_info()
```
